```python
import numpy as np
import pandas as pd
from PIL import Image
import os
```

```python
df = pd.read_csv('datasets/test.csv')
```

```python
df.sample(5)
```

```python
os.makedirs('static/test_images' , exist_ok=True)
```

```python
for index , row in df.iterrows():
    img_array = np.array(row).reshape(28,28).astype('uint8')
    img = Image.fromarray(img_array)
    img.save(f'static/test_images/img_{index}.png')



print("✅ Done! All 28,000 images saved to static/test_images/")
```
